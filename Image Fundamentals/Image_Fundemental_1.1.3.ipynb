```python

import math
import matplotlib.pyplot as plt
import cv2

picture  = cv2.imread("Barbara.bmp")
picture  = cv2.cvtColor(picture, cv2.COLOR_BGR2GRAY)
plt.imshow(picture)
```

```python
def binaryToDecimal(binary , number):
    num = ""
    for i in range(number):
        num = num + binary[i]
    for j in range( number , 8):
        num = num +'0'
    int_num = int(num)
    binary1 = int_num
    decimal, i, n = 0, 0, 0
    while(int_num != 0):
        dec = int_num % 10
        decimal = decimal + dec * pow(2, i)
        int_num = int_num//10
        i += 1
    return decimal
```

```python
import numpy as np
x , y = picture.shape
pixels = np.zeros([x , y ])
```

```python
 for i in range(x) :
        for j in range(y) :
#             here we convert pixels from int to 8bits binary
            pixels[i][j] = format(picture[i][j] , '08b') 
            
```
